```python
import whisper

# STEP 1: Transcribe
model = whisper.load_model("base")
result = model.transcribe("12thJul2025.mov")
transcript = result['text']

print(transcript)
```

```python
print("Transcription complete.", "Transcript length:", len(transcript), "characters.")
# need words length
print("Transcription complete.", "Words length:", len(transcript.split()), "words.")

```

```python
import ollama  # This import is not used in the current context, but kept for reference.

# STEP 2: Generate Summary
# Assuming you have a model named 'summarize' in Ollama
summary = ollama.chat(
    model="phi3:mini",
    messages=[
        {"role": "user", "content": f"tell me funny parts only:\n{transcript}"}
    ]
)

print(summary['message']['content'])
```
