```python
import noisereduce as nr
import librosa
import soundfile as sf

# Load audio
y, sr = librosa.load("louder.wav", sr=None)

# Reduce noise (simple version)
reduced_noise = nr.reduce_noise(y=y, sr=sr)

# Save cleaned audio
sf.write("output_clean.wav", reduced_noise, sr)
```

```python
from pydub import AudioSegment

# Load audio from video
video = AudioSegment.from_file("16thJul2025.mov", format="mov")

# Increase volume
louder_audio = video + 40  # increase volume by 15 dB

# Export to wav
louder_audio.export("louder.wav", format="wav")
```
